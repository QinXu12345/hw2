```python
from hw2 import Uniform,Exponential,Xnorgenerator,Normal
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use("ggplot")
SAMPLES = 100000
BINS = 100
KS_SAMPLES = 1000
```

```python
normal = Normal(mean=np.zeros(2),cov=np.array([[1,2],[2,5]]))
```

```python
samples = np.array([normal.rvs() for _ in range(SAMPLES)])
```

```python
fig,ax = plt.subplots()
sns.scatterplot(x=samples[:,0],y=samples[:,1],ax=ax)
ax.set_title("Point Plot of Our Normal distribution")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
norm = np.sqrt(samples[:,0] ** 2 + samples[:,1] ** 2)
norm = np.expand_dims(norm, axis=1)
angles = samples / norm
angles = np.arctan2(angles[:,1],angles[:,0]) * 180 / np.pi
angles = angles + 180
angles
```

```python
theta = np.linspace(0,2*np.pi,10000)
f_theta = 1/(2*np.pi *(2*np.cos(theta)-np.sin(theta))**2 + np.cos(theta)**2)
```

```python
fig,axs = plt.subplots(ncols=2,figsize=(15,6))
sns.histplot(angles,binwidth=0.1,kde=True,ax=axs[0],stat="density")
sns.lineplot(x=theta * 180/np.pi,y=f_theta,ax=axs[1])
axs[0].set_title("Sample distribution of $\\theta$")
axs[1].set_title("Theoretical Value of $\\theta$")
axs[0].set_xlabel("$\\theta$ (0,2$\\pi$)")
axs[1].set_xlabel("$\\theta$ (0,2$\\pi$)")
# plt.sup_title("Theoretical value vs Sample value")
```
